# Exercício - Iteração em um Numpy array

```python
import numpy as np
```

```python
números = np.array([5.5, 3.8, 9, 7.5, 10.0, 9.9, 8.5])
```

```python
valores = [5.5, 3.8, 9, 7.5, 10.0, 9.9, 8.5]
```

```python
# Calcule a média dos valores usando iteração

print(np.average(valores))
```

```python
# Calcule o desvio padrão dos valores usando iteração.

print(np.std(valores))
```

```python
# Ordene os elementos do array da questão anterior usando o método .sort() próprio do numpy.

sorted_array = np.sort(valores)
print(sorted_array)
```

```python
#  Para saber as dimensões do array podemos use a propriedade .shape

dimensions = np.shape(valores)
print(dimensions)
```

```python
# Extraia a média de um array numpy utilizando o método .mean():

print(np.mean(valores))
```

```python
#O mesmo pode ser feito para o desvio padrão, com .std(), máxima e mínima com .max() e .min(), além de muitos outros métodos.

print(np.std(valores))

print(np.max(valores))

print(np.min(valores))
```
